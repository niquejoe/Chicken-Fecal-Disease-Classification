```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

# Preprocess Your Data (Do this once)

```python
import os
from os import listdir
import csv
import shutil
```

```python
fecal_folder = "Train"
filename = 'train_data.csv'

for images in os.listdir(fecal_folder):
    if (images.endswith(".jpg")):
        with open(filename, 'r') as csvfile:
            datareader = csv.reader(csvfile)
            for row in datareader:
                if(images == row[0]): 
                    shutil.copy("Train/"+images, "Datasets/"+row[1])
                    print(images + " Copied to: " +row[1])
```

# File Access

```python
import pathlib
data_dir = pathlib.Path("Datasets")
```

```python
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

# Create Data Pipeline

```python
batch_size = 32
img_height = 180
img_width = 180
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2, #meaning 80% will be used for training, 20% for testing
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

# Start Training

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
normalization_layer = layers.Rescaling(1./255)
normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
```

```python
num_classes = len(class_names)
```

# preparing your model

```python
model = Sequential([
  layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary() 
```

# Model Training and Generation

```python
epochs=100 #how many cycles of training () suggest to be 100
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

# Try with real image

```python
test_path = "Test/pcrsalmo.1.jpg"
```

```python
img = tf.keras.utils.load_img(
    test_path, target_size=(img_height, img_width)
)
img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch
```

```python
predictions = model.predict(img_array)
```

```python
score = tf.nn.softmax(predictions[0])
```

```python
print("")
print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(score)], 100 * np.max(score))
)
```
